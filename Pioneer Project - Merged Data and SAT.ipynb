# EDA
## Bivariate Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
plt.scatter(df['D'], df['Average Total'])
plt.title('Average SAT Scores vs D Subcategory')
plt.xlabel('D Subcategory Scores')
plt.ylabel('Average Total SAT Scores')
plt.show()
```

```python
plt.scatter(df['E'], df['Average Total'])
plt.title('Average SAT Scores vs E Subcategory')
plt.xlabel('E Subcategory Scores')
plt.ylabel('Average Total SAT Scores')
plt.show()
```

```python
plt.scatter(df['G'], df['Average Total'])
plt.title('Average SAT Scores vs G Subcategory')
plt.xlabel('G Subcategory Scores')
plt.ylabel('Average Total SAT Scores')
plt.show()
```

```python
plt.scatter(df['CL'], df['Average Total'])
plt.title('Average SAT Scores vs Civil Liberties')
plt.xlabel('Total Civil Liberties Scores')
plt.ylabel('Average Total SAT Scores')
plt.show()
```

```python
plt.scatter(df['Total'], df['Average Total'])
plt.title('Average SAT Scores vs Total Freedom')
plt.xlabel('Total Freedom Scores')
plt.ylabel('Average Total SAT Scores')
plt.show()
```

## Multivariate Analysis

```python
import numpy as np
%matplotlib inline

plt.figure(figsize=(10,5))
var = ['Math', 'Reading', 'Average Total', 'CL', 'D', 'E', 'G']
data = df[var]
sns.heatmap(data.corr(), annot = True)
plt.title('Correlation Map of All Variables')
```
