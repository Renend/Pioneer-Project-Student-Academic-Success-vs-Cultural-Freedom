```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import style
```

```python
plt.scatter(df['E'], df['Average'])
plt.title('Average Gaokao vs E Subcategory Scores')
plt.xlabel('E Subcategory Scores')
plt.ylabel('Average Gaokao Score/Admission Lines')
plt.show()
```

```python
plt.scatter(df['G'], df['Average'])
plt.title('Average Gaokao vs G Subcategory Scores')
plt.xlabel('G Subcategory Scores')
plt.ylabel('Average Gaokao Score/Admission Lines')
plt.show()
```
